<a id="glossary"></a>
<h2>üìå Glossary</h2>

<details>
<summary><strong>üìñ Click to Expand</strong></summary>

<h5>üß† Basic Concepts</h5>

<ul>
  <li>üé≤ <strong>Random Variable</strong>: A variable whose values depend on outcomes of a random phenomenon. Usually denoted by <code>X</code>. <code>X</code> can be <strong>Discrete</strong> or <strong>Continuous</strong>.</li>
  <li>üóÇÔ∏è <strong>Probability Distribution</strong>: Describes the probability values for each possible value a random variable can take.
    <ul>
      <li>For <strong>discrete</strong> variables: <strong>Probability Mass Function (PMF)</strong></li>
      <li>For <strong>continuous</strong> variables: <strong>Probability Density Function (PDF)</strong></li>
    </ul>
  </li>
  <li>üìà <strong>Cumulative Distribution Function (CDF)</strong>: Describes the probability that a variable takes a value <strong>less than or equal to</strong> a certain value. Represents total probability from <code>X = -‚àû</code> to <code>X = x</code>.</li>
  <li>üìä <strong>Expected Value</strong> (ùîº[X]): The long-run average value a random variable is expected to take. Also called the <strong>Expectation</strong> or <strong>Mean</strong>.</li>
  <li>üìâ <strong>Variance</strong> (Var(X)): Measures spread around the mean. Higher variance means more dispersion.</li>
  <li>üìè <strong>Standard Deviation</strong> (œÉ): Square root of variance ‚Äî shows typical deviation from the mean.</li>
</ul>

</details>


<a id="types-of-random-variables"></a>
<h2>üé≤ Types of Random Variables</h2>

<details>
<summary><strong>üìñ Click to Expand</strong></summary>

<ul>
  <li>üî¢ <strong>Discrete Random Variable</strong>: A variable that can take a <strong>countable</strong> number of values (e.g., number of customers in a store).<br>
    üß™ Example: Binomial, Poisson distributions.
  </li>

  <li>üìè <strong>Continuous Random Variable</strong>: A variable that can take <strong>infinitely many values</strong> within a range (e.g., height, weight).<br>
    üß™ Example: Normal, Exponential distributions.
  </li>
</ul>

</details>


<a id="probability"></a>
<h2>üéØ Probability</h2>

<details>
<summary><strong>üìñ Click to Expand</strong></summary>

<ul>
  <li>üîó <strong>Independence</strong>: Two events are independent if the outcome of one <strong>does not</strong> affect the outcome of the other.</li>

  <li>üéØ <strong>Conditional Probability</strong> (<code>P(A | B)</code>): The probability of event <code>A</code> occurring <strong>given that</strong> event <code>B</code> has already occurred.</li>

  <li>üîÑ <strong>Bayes‚Äô Theorem</strong>: A formula used to update probabilities based on new information. Given prior probability <code>P(A)</code> and likelihood <code>P(B | A)</code>, it calculates the posterior probability <code>P(A | B)</code>:
    <pre>
P(A | B) = (P(B | A) ¬∑ P(A)) / P(B)
    </pre>
  </li>

  <li>üìà <strong>Law of Large Numbers (LLN)</strong>: As the number of trials increases, the sample mean <strong>approaches</strong> the true population mean.</li>

  <li>üßÆ <strong>Central Limit Theorem (CLT)</strong>: The sum or average of a <strong>large</strong> number of independent random variables follows a <strong>Normal Distribution</strong>, regardless of the original distribution.</li>
</ul>

</details>


<a id="estimation-inference"></a>
<h2>üß† Estimation & Inference</h2>

<details>
<summary><strong>üìñ Click to Expand</strong></summary>

<ul>
  <li>üìà <strong>Maximum Likelihood Estimation (MLE)</strong>: A method to estimate parameters by maximizing the likelihood of observed data.</li>
  <li>üßÆ <strong>Method of Moments (MoM)</strong>: A parameter estimation technique that equates sample moments to theoretical moments.</li>
  <li>üìä <strong>Bayesian Inference</strong>: A method of statistical inference that updates prior beliefs based on observed data.</li>
  <li>üéØ <strong>Confidence Interval (CI)</strong>: A range of values that is likely to contain the true parameter value with a certain probability (e.g., 95% CI).</li>
  <li>üìâ <strong>p-value</strong>: The probability of observing the given data (or something more extreme) if the null hypothesis is true. A small p-value (&lt; 0.05) suggests strong evidence <strong>against</strong> the null hypothesis.</li>
  <li>üß™ <strong>Hypothesis Testing</strong>: A process to determine if a hypothesis is statistically significant.
    <ul>
      <li>‚ö™ <strong>Null Hypothesis (H‚ÇÄ)</strong>: Assumes no effect or no difference.</li>
      <li>üî¥ <strong>Alternative Hypothesis (H‚ÇÅ)</strong>: Assumes a significant effect or difference.</li>
    </ul>
  </li>
</ul>

</details>

<a id="uniform"></a>
<h2>üìä Uniform</h2>

<details>
<summary><strong>üìñ Click to Expand</strong></summary>

<h5>üé≤ What is a Uniform Distribution?</h5>

<p>In a <strong>Uniform distribution</strong>, every value in a certain range is equally likely to occur. There‚Äôs no bias toward any number ‚Äî it‚Äôs perfectly "fair."</p>

<h5>üìå Parameters</h5>

<ul>
  <li><code>a</code>: the <strong>minimum</strong> value (start of the range)</li>
  <li><code>b</code>: the <strong>maximum</strong> value (end of the range)</li>
  <li>The distribution is flat between <code>a</code> and <code>b</code></li>
</ul>

<h5>üß† Real-Life Examples</h5>

<ul>
  <li>üé≤ <strong>Rolling a fair die</strong>: Each face (1‚Äì6) has equal chance ‚Äî this is a discrete uniform.</li>
  <li>üî¢ <strong>Random number generator</strong>: Picking a number between 0 and 1 (or 0 and 10), where every number is equally likely.</li>
  <li>üéØ <strong>Landing randomly on a line</strong>: If a dart lands randomly on a 10-meter line, it's equally likely to hit any point between 0 and 10.</li>
</ul>

<h5>üßæ When to Use</h5>

<p>Use the Uniform distribution when you know a range but <strong>don‚Äôt expect any value to be more likely than others</strong> ‚Äî complete uncertainty within bounds.</p>

</details>


#### Sample from Distribution

```python
import numpy as np

# Define the range [a, b]
a, b = 0, 10
size = 10000  

# Generate random samples from the Uniform distribution
samples = np.random.uniform(a, b, size)

# Print the first 10 samples as a quick check
print(np.round(samples[:10], 2))
```

#### Key Properties

```python
import scipy.stats as stats

# 1. Probability Density Function (PDF)
x = np.linspace(min(samples), max(samples), 100)
pdf_values = stats.uniform.pdf(x, a, b-a)

# 2. Cumulative Distribution Function (CDF)
cdf_values = stats.uniform.cdf(x, a, b-a)

# 3. Print the values
np.set_printoptions(edgeitems=10, threshold=20, precision=6)
print("x values:", x, "\n")
print("PDF values:", pdf_values, "\n")
print("CDF values:", cdf_values, "\n")
```

#### Summarize Distribution

<details>
<summary><strong>üìñ Click to Expand</strong></summary>

<ul>
  <li><strong>üìå Mean</strong>: The average of all values. Gives a sense of the central tendency.</li>
  <li><strong>üìà Variance</strong>: How spread out the values are. Higher variance means more variability.</li>
  <li><strong>üìè Std Dev (Standard Deviation)</strong>: Like variance, but in the same units as the data. Easier to interpret.</li>
  <li><strong>üîÄ Skewness</strong>: Tells you if the data is asymmetric.
    <ul>
      <li>Positive = longer right tail (more high values).</li>
      <li>Negative = longer left tail (more low values).</li>
    </ul>
  </li>
  <li><strong>üéØ Kurtosis</strong>: Measures "tailedness".
    <ul>
      <li>High kurtosis = more outliers.</li>
      <li>Low kurtosis = flatter distribution.</li>
    </ul>
  </li>
</ul>

</details>


<details>
<summary><strong>üìñ Click to Expand</strong></summary>

<p>In simple terms, <strong>moments</strong> describe the <em>shape</em> of a distribution beyond just averages and variability:</p>

<ul>
  <li><strong>1st Moment (Mean)</strong>: The average ‚Äî where the data tends to center.</li>
  <li><strong>2nd Moment (Variance)</strong>: How spread out the data is around the mean.</li>
  <li><strong>3rd Moment (Skewness)</strong>: The tilt or asymmetry of the distribution.
    <ul>
      <li>Positive skew ‚Üí long tail on the right</li>
      <li>Negative skew ‚Üí long tail on the left</li>
    </ul>
  </li>
  <li><strong>4th Moment (Kurtosis)</strong>: How heavy or light the tails are.
    <ul>
      <li>High kurtosis ‚Üí more outliers</li>
      <li>Low kurtosis ‚Üí flatter shape</li>
    </ul>
  </li>
</ul>

<p>These help us understand not just the <em>average behavior</em>, but the <em>risk and irregularities</em> in the data.</p>

</details>


```python
def summarize_distribution(samples):
    """
    Compute and print basic statistics of a given sample.
    
    Parameters:
    - samples (array-like): Sample data from a distribution.
    """
    stats_dict = {
        "üìå Mean": round(np.mean(samples), 4),
        "üìà Variance": round(np.var(samples), 4),
        "üìè Std Dev": round(np.std(samples), 4),
        "üîÄ Skewness": round(stats.skew(samples), 4),
        "üéØ Kurtosis": round(stats.kurtosis(samples), 4),
    }

    for label, value in stats_dict.items():
        print(f"{label:<12}: {value:>8.4f}")
```

```python
summarize_distribution(samples)
```

```python
def summarize_moments(samples, raw=True):
    """
    Compute and print the 1st through 4th moments of a sample.

    Parameters:
    - samples (array-like): Sample data from a distribution.
    - raw (bool): If True, compute raw moments (E[X^k]); if False, compute central moments.
    """
    if raw:
        moments = [np.mean(samples**i) for i in range(1, 5)]
        label = "Raw"
    else:
        moments = [stats.moment(samples, moment=i) for i in range(1, 5)]
        label = "Central"

    print(f"üìä {label} Moments:")
    for i, val in enumerate(moments, start=1):
        print(f"  {i}·µó ∞ Moment: {val:.4f}")
```

```python
summarize_moments(samples)
```

#### Visualizing Distributions

<details>
<summary><strong>üìñ Click to Expand</strong></summary>

<p>A <strong>QQ-Plot (Quantile-Quantile Plot)</strong> is a visual way to answer this question:</p>

<blockquote>
<p><em>"Does my data look like it came from a normal distribution?"</em></p>
</blockquote>

<h5>üîç Purpose</h5>
<ul>
  <li>It compares your <strong>sample data</strong> to a <strong>perfect normal distribution</strong>.</li>
  <li>If your data <strong>follows a straight line</strong>, it's likely normal.</li>
  <li>Deviations from the line = skewness, heavy tails, or outliers.</li>
</ul>

<h5>üß† Interpretation</h5>
<ul>
  <li><strong>Linearity</strong>: Points closely follow the diagonal ‚Üí roughly normal.</li>
  <li><strong>S-shape</strong>: Suggests skewness.</li>
  <li><strong>Heavy tails</strong>: Points bend away from the ends of the line.</li>
</ul>

<h5>‚öôÔ∏è What does the function do?</h5>
<ul>
  <li>It <strong>sorts your sample</strong> and plots its quantiles.</li>
  <li>It compares them to the <strong>expected quantiles from a normal distribution</strong>.</li>
  <li>The red line is the theoretical ‚Äúperfect normal‚Äù reference.</li>
  <li>The function uses <code>scipy.stats.probplot()</code> internally.</li>
</ul>

<p>This is a quick visual <strong>sanity check</strong> for normality before running tests or models that assume normal data.</p>

</details>


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns

def plot_distribution(samples, distribution_name, dist_func=None, params=None, discrete=False):
    """
    General function to visualize a distribution.
    
    Args:
        samples (np.array): The generated random samples.
        distribution_name (str): Name of the distribution (for title/labels).
        dist_func (scipy.stats distribution, optional): Distribution function from scipy.stats (e.g., stats.norm).
        params (tuple, optional): Parameters for the distribution function (e.g., (mean, std) for normal).
        discrete (bool): Set to True for discrete distributions.
    """

    fig, axs = plt.subplots(1, 3, figsize=(18, 5))
    
    # 1Ô∏è‚É£ Histogram vs. Theoretical Distribution (PMF/PDF)
    axs[0].hist(samples, bins=50 if not discrete else np.arange(min(samples), max(samples) + 1.5) - 0.5,
                density=True, alpha=0.6, color='blue', edgecolor='black', label="Sampled Data (Observed)")
    
    if dist_func and params:
        x = np.linspace(min(samples), max(samples), 100) if not discrete else np.arange(min(samples), max(samples) + 1)
        if discrete:
            y = dist_func.pmf(x, *params)
        else:
            y = dist_func.pdf(x, *params)
        axs[0].plot(x, y, 'r-', label="Theoretical")
    
    axs[0].set_title(f"{distribution_name} - Actual vs. Theoretical")
    axs[0].set_xlabel("Value")
    axs[0].set_ylabel("Probability Density/Mass")
    axs[0].legend()

    # 2Ô∏è‚É£ CDF Plot
    sorted_samples = np.sort(samples)
    empirical_cdf = np.arange(1, len(sorted_samples) + 1) / len(sorted_samples)
    axs[1].plot(sorted_samples, empirical_cdf, marker="o", linestyle="none", label="Empirical CDF")
    
    if dist_func and params:
        theoretical_cdf = dist_func.cdf(sorted_samples, *params)
        axs[1].plot(sorted_samples, theoretical_cdf, 'r-', label="Theoretical CDF")
    
    axs[1].set_title(f"{distribution_name} - CDF")
    axs[1].set_xlabel("Value of Random Variable X")
    axs[1].set_ylabel("Cumulative Probability")
    axs[1].legend()

    # 3Ô∏è‚É£ QQ-Plot (for normality check)
    stats.probplot(samples, dist="norm", plot=axs[2])
    axs[2].set_title(f"{distribution_name} - QQ-Plot")

    plt.tight_layout()
    plt.show()
```

```python
plot_distribution(samples, "Uniform Distribution", stats.uniform, (0, 10))
```

<h4>Parameter Estimation</h4>

<details>
<summary><strong>üìñ Click to Expand</strong></summary>

Parameter Estimation is opposite process as Sampling:

<ul>
    <li> Sampling: You start with a known distribution and its parameters to generate random data. </li>
    <li> Parameter Estimation: You start with observed data and an assumed distribution to estimate the parameters. </li>
</ul>

</details>


<p><code>Maximum Likelihood Estimation (MLE)</code></p>

<details>
<summary><strong>üìñ Click to Expand</strong></summary>

<p><strong>Maximum Likelihood Estimation (MLE)</strong> finds the parameter values that make the observed data <strong>most likely</strong>.</p>

<p>For a <strong>Uniform distribution</strong>, the best guess for the minimum (<code>a</code>) and maximum (<code>b</code>) is just the smallest and largest observed values. Why? Because including more extreme values would make the data less "likely" under the assumed uniform model.</p>

<p>üîπ <strong>How it works:</strong></p>
<ul>
  <li>Assumes the data came from a uniform distribution.</li>
  <li>Picks <code>a = min(data)</code> and <code>b = max(data)</code> to maximize the likelihood.</li>
</ul>

</details>


```python
def mle_uniform(samples):
    """ MLE for Uniform Distribution: Estimates a (min) and b (max) """
    estimated_a = np.min(samples)
    estimated_b = np.max(samples)
    return estimated_a, estimated_b

# Example usage
estimated_a, estimated_b = mle_uniform(samples)
print(f"MLE Estimated a: {estimated_a:.3f}, b: {estimated_b:.3f}")
```

<p><code>Method of Moments (MoM)</code></p>

<details>
<summary><strong>üìñ Click to Expand</strong></summary>

<p><strong>Method of Moments (MoM)</strong> estimates parameters by matching sample statistics (like mean and variance) to their theoretical values.</p>

<p>For a <strong>Uniform distribution</strong>:</p>
<ul>
  <li>Theoretical mean = (a + b) / 2</li>
  <li>Theoretical variance = (b ‚àí a)¬≤ / 12</li>
</ul>

<p>MoM solves these equations using the sample mean and variance to estimate <code>a</code> and <code>b</code>.</p>

<p>üîπ <strong>How it works:</strong></p>
<ul>
  <li>Compute sample mean and variance.</li>
  <li>Solve two equations to estimate <code>a</code> and <code>b</code>.</li>
</ul>

</details>


```python
def mom_uniform(samples):
    """ MoM for Uniform Distribution: Estimates a (min) and b (max) """
    sample_mean = np.mean(samples)
    sample_var = np.var(samples)

    estimated_a = sample_mean - np.sqrt(3 * sample_var)
    estimated_b = sample_mean + np.sqrt(3 * sample_var)
    
    return estimated_a, estimated_b

# Example usage
estimated_a, estimated_b = mom_uniform(samples)
print(f"MoM Estimated a: {estimated_a:.3f}, b: {estimated_b:.3f}")
```

<p><code>Bayesian Inference</code></p>

<details>
<summary><strong>üìñ Click to Expand</strong></summary>

<p><strong>Bayesian Estimation</strong> combines <strong>prior beliefs</strong> with observed data to estimate parameters.</p>

<p>Here, we assume very broad beliefs about where <code>a</code> and <code>b</code> might be (e.g., between 0 and 20), and use the data to update that belief. For simplicity, this example just uses the <strong>MAP estimate</strong>, which ends up being the min and max of the data ‚Äî similar to MLE, especially with weak priors.</p>

<p>üîπ <strong>How it works:</strong></p>
<ul>
  <li>Start with prior assumptions (e.g., uniform priors).</li>
  <li>Use data to update those assumptions.</li>
  <li>Output the most probable values (MAP estimate).</li>
</ul>

</details>


```python
import scipy.stats as stats

def bayesian_uniform(samples, prior_range=(0, 20)):
    """ Bayesian estimation for Uniform Distribution using weakly informative priors """
    prior_a = stats.uniform(0, prior_range[1])  # Prior for a
    prior_b = stats.uniform(0, prior_range[1])  # Prior for b

    estimated_a = np.min(samples)  # Approximate MAP estimate
    estimated_b = np.max(samples)

    return estimated_a, estimated_b

# Example usage
estimated_a, estimated_b = bayesian_uniform(samples)
print(f"Bayesian Estimated a: {estimated_a:.3f}, b: {estimated_b:.3f}")
```

<h2>üìà Normal<br><span style="font-weight:normal;">(AKA Gaussian)</span></h2>

<details>
<summary><strong>üìñ Click to Expand</strong></summary>

<h5>üìà What is a Normal Distribution?</h5>

<p>The <strong>Normal distribution</strong> (also called Gaussian) is the classic bell curve ‚Äî values cluster around a central mean, with fewer values as you move further out.</p>

<h5>üìå Parameters</h5>

<ul>
  <li><code>Œº</code> (mu): the <strong>mean</strong> ‚Äî where the center of the curve is</li>
  <li><code>œÉ</code> (sigma): the <strong>standard deviation</strong> ‚Äî controls how wide or narrow the curve is</li>
</ul>

<h5>üß† Real-Life Examples</h5>

<ul>
  <li>üéì <strong>Test scores</strong>: In large exams, most students score around the average, with fewer scoring very low or very high.</li>
  <li>‚öñÔ∏è <strong>Heights or weights</strong>: Adult human heights tend to follow a bell-shaped distribution.</li>
  <li>üè≠ <strong>Manufacturing errors</strong>: Small random variations in machine production tend to follow a normal curve.</li>
</ul>

<h5>üßæ When to Use</h5>

<p>Use the Normal distribution when data tends to <strong>cluster around a central value</strong> and the deviations are symmetric in both directions.</p>

</details>


#### Sample from Distribution

#### Parameter Estimation

`Maximum Likelihood Estimation (MLE)`

```python
from scipy.optimize import minimize

def mle_normal(samples):
    """ MLE for Normal Distribution: Estimates mean and std deviation """
    def neg_log_likelihood(params):
        mu, sigma = params
        return -np.sum(stats.norm.logpdf(samples, mu, sigma))
    
    init_params = [np.mean(samples), np.std(samples)]
    result = minimize(neg_log_likelihood, init_params, method="L-BFGS-B")
    return result.x  # [estimated_mu, estimated_sigma]

# Example usage
estimated_mu, estimated_sigma = mle_normal(samples)
print(f"MLE Estimated Œº: {estimated_mu:.3f}, œÉ: {estimated_sigma:.3f}")
```

`Method of Moments (MoM)`

```python
def mom_normal(samples):
    """ MoM for Normal Distribution: Estimates mean and std deviation """
    estimated_mu = np.mean(samples)
    estimated_sigma = np.std(samples)
    return estimated_mu, estimated_sigma

# Example usage
estimated_mu, estimated_sigma = mom_normal(samples)
print(f"MoM Estimated Œº: {estimated_mu:.3f}, œÉ: {estimated_sigma:.3f}")
```

`Bayesian Inference`

```python
def bayesian_normal(samples, prior_mu=0, prior_sigma=10):
    """ Bayesian estimation for Normal Distribution using Normal-Gamma prior """
    n = len(samples)
    sample_mean = np.mean(samples)
    sample_var = np.var(samples)

    posterior_mu = (prior_mu + n * sample_mean) / (n + 1)
    posterior_sigma = np.sqrt(sample_var / (n + 1))

    return posterior_mu, posterior_sigma

# Example usage
estimated_mu, estimated_sigma = bayesian_normal(samples)
print(f"Bayesian Estimated Œº: {estimated_mu:.3f}, œÉ: {estimated_sigma:.3f}")
```

```python
import numpy as np

# Define degrees of freedom
df = 10  
size = 10000  

# Generate samples
samples = np.random.standard_t(df, size)

# Print first 10 samples
print(np.round(samples[:10], 2))

```

#### Key Properties

```python
# Define x range
x = np.linspace(min(samples), max(samples), 100)

# Degrees of freedom (df) estimated from sample variance
df = len(samples) - 1

# 1. Probability Density Function (PDF)
pdf_values = stats.t.pdf(x, df)

# 2. Cumulative Distribution Function (CDF)
cdf_values = stats.t.cdf(x, df)

# 3. Print the values
np.set_printoptions(edgeitems=10, threshold=20, precision=6)
print("x values:", x, "\n")
print("PDF values:", pdf_values, "\n")
print("CDF values:", cdf_values, "\n")
```

#### Parameter Estimation

`Maximum Likelihood Estimation (MLE)`

```python
def mle_student_t(samples):
    """ MLE for Student's t-Distribution: Estimates degrees of freedom (df) """
    def neg_log_likelihood(df):
        return -np.sum(stats.t.logpdf(samples, df))
    
    result = minimize(neg_log_likelihood, x0=[10], method="L-BFGS-B", bounds=[(1, None)])
    return result.x[0]

# Example usage
estimated_df = mle_student_t(samples)
print(f"MLE Estimated df: {estimated_df:.3f}")
```

`Method of Moments (MoM)`

```python
def mom_student_t(samples):
    """ MoM for Student's t-Distribution: Estimates degrees of freedom (df) """
    sample_var = np.var(samples)
    
    def solve_df(df):
        return df / (df - 2) - sample_var  # Equation to solve

    from scipy.optimize import fsolve
    estimated_df = fsolve(solve_df, x0=10)[0]
    return estimated_df

# Example usage
estimated_df = mom_student_t(samples)
print(f"MoM Estimated df: {estimated_df:.3f}")
```

`Bayesian Inference`

```python
def bayesian_student_t(samples, alpha_prior=1, beta_prior=1):
    """ Bayesian estimation for Student's t-Distribution using Gamma prior """
    n = len(samples)
    posterior_alpha = alpha_prior + n / 2
    posterior_beta = beta_prior + np.sum(samples**2) / 2

    return posterior_alpha, posterior_beta

# Example usage
posterior_alpha, posterior_beta = bayesian_student_t(samples)
print(f"Bayesian Estimated df ~ Gamma({posterior_alpha}, {posterior_beta:.3f})")
```
