```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm  # Example with normal distribution

def manual_qqplot(sample, dist_ppf):
    sample_sorted = np.sort(sample)
    n = len(sample_sorted)
    probs = [(i - 0.5) / n for i in range(1, n + 1)]
    theoretical_quantiles = [dist_ppf(p) for p in probs]
    return theoretical_quantiles, sample_sorted
```

```python
# Example usage:
# Replace this with your actual sample
np.random.seed(0)
sample = np.random.uniform(0, 10, 1000)

# For plotting against normal distribution
x, y = manual_qqplot(sample, uniform(loc=0, scale=10).ppf)
```

```python
# Plot
plt.scatter(x, y, color='blue', s=5)
plt.plot(x, x, color='red')  # Reference line
plt.title("Uniform Distribution - QQ-Plot")
plt.xlabel("Theoretical quantiles")
plt.ylabel("Ordered Values")
plt.grid(True)
plt.show()
```
